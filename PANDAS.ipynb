```python
import pandas as pd
import numpy as np

cars = pd.Series(np.random.choice(["BMW", "HONDA", "Nissan"], 10))
colors = pd.Series(np.random.choice(["red", "black", "blue", "white"], 10))
odometer = pd.Series(np.random.randint(1000, 200000, 10))
doors = pd.Series(np.random.choice([2, 3, 4, 5], 10))
price = pd.Series(np.random.randint(50000, 2000000, 10))

CAR_SLAES = {
    "Car": cars,
    "Color": colors,
    "Odometer": odometer,
    "Doors": doors,
    "Price": price
}

df = pd.DataFrame(CAR_SLAES )

df.to_csv("cars_data.csv", index=False)
print(df)
```
