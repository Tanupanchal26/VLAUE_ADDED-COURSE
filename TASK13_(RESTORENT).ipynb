```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))

ax=sns.countplot(x='sex',data=df,palette='cool',edgecolor='black')

# Add title and labels
ax.set_title("sex Distribution", fontsize=14, fontweight='bold')
ax.set_xlabel("sex", fontsize=12)
ax.set_ylabel("Count", fontsize=12)

# Add data labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

sns.despine()
plt.tight_layout()
plt.show()
```

```python
sales_gen = df.groupby(['sex'], as_index=False)['total_bill'].sum().sort_values (by='total_bill', ascending=False)

sns.barplot(x = 'sex', y= 'total_bill', data = sales_gen)
plt.show()
```

From above graphs we can see that male customers contribute a higher total bill amount than female customers.

```python
ax = sns.countplot(
    data=df,x="sex", hue="sex", palette={"Female": "pink", "Male": "blue"},edgecolor='black')

plt.show()
```

```python
 # Total Amount vs Age Group

sales_age = df.groupby(['sex'], as_index=False)['total_bill'].sum().sort_values(by='total_bill', ascending=False)
sns.barplot(x = 'sex', y= 'total_bill', data = sales_age)


```

From above graphs we can see that  male buyers have higher total spending than female buyers.

```python
orders_state = df.groupby(['day'], as_index=False)['total_bill'].sum()\
                 .sort_values(by='total_bill', ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,4)})

ax = sns.barplot(x='day', y='total_bill', data=orders_state, palette='rocket')

plt.title("Top Days by Total Bill", fontsize=14, fontweight='bold')
plt.xlabel("Day", fontsize=12)
plt.ylabel("Total Bill", fontsize=12)
plt.xticks(rotation=60)

# Add labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.tight_layout()
plt.show()
```

```python
sales_state = df.groupby (['day'], as_index=False) ["total_bill"].sum().sort_values (by='total_bill', ascending=False).head(10)

sns.set(rc={'figure.figsize': (15,5)})

sns.barplot(data = sales_state, x = 'day', y= 'total_bill')

plt.xticks(rotation=30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", rc={'figure.figsize': (8,4)})

ax = sns.histplot(data=df, x='tip', bins=20, kde=True, color='skyblue', edgecolor='black')

plt.title("Distribution of Tips", fontsize=12, fontweight='bold')
plt.xlabel("Tip Amount")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sales_state = df.groupby(['tip_round', 'sex'], as_index=False)['total_bill'].sum() \
                .sort_values(by='total_bill', ascending=False)

sns.set(style="whitegrid", rc={'figure.figsize': (8,5)})

ax = sns.barplot(data=sales_state, x='tip_round', y='total_bill', hue='sex', palette='deep')

plt.title("Total Bill by Tip and Gender", fontsize=13, fontweight='bold')
plt.xlabel("Tip Amount")
plt.ylabel("Total Bill")

# Add labels
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

From above graphs we can see that ale customers have higher total spending than female customers across different tip amounts.

```python
sales_state = df.groupby(['size'], as_index=False)['total_bill'].sum().sort_values(by='total_bill', ascending=False).head(10)

sns.set(rc={'figure.figsize': (10,5)})

sns.barplot(data=sales_state, x='size', y='total_bill')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sales_state = df.groupby(['sex'], as_index=False).size() \
                .sort_values(by='size', ascending=False)

sns.set(rc={'figure.figsize': (6,4)})
ax = sns.barplot(data=sales_state, x='sex', y='size', palette='deep')

# Add labels
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```
