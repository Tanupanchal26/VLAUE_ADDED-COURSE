# 1. NumPy (Numerical Python)

Purpose: Efficient numerical computations, especially with arrays and matrices.

------------------

# Why we need it:

Python lists are slow for math operations on large data. NumPy provides arrays that are fast, memory-efficient, and support vectorized operations (do many calculations at once).

----------------------------

# Key feature:

Multi-dimensional arrays (ndarray)
Fast math operations like addition, multiplication, mean, sum, etc.


----------------------------------
| Library   | Category           | Function / Command           | Description / Formula       |
| --------- | ------------------ | ---------------------------- | --------------------------- |
| **NumPy** | Array Creation     | `np.array([1,2,3])`          | Create 1D array             |
|           |                    | `np.array([[1,2],[3,4]])`    | Create 2D array             |
|           |                    | `np.zeros(5)`                | Array of zeros              |
|           |                    | `np.ones(3)`                 | Array of ones               |
|           |                    | `np.arange(0,10,2)`          | Array from 0 to 10 step 2   |
|           |                    | `np.linspace(0,1,5)`         | 5 numbers evenly spaced 0–1 |
|           | Math Ops           | `a + b, a - b, a * b, a / b` | Element-wise arithmetic     |
|           |                    | `a ** 2, np.sqrt(a)`         | Square, square root         |
|           | Aggregation        | `np.sum(arr)`                | Sum of array                |
|           |                    | `np.mean(arr)`               | Mean                        |
|           |                    | `np.median(arr)`             | Median                      |
|           |                    | `np.std(arr)`                | Standard deviation          |
|           |                    | `np.var(arr)`                | Variance                    |
|           |                    | `np.min(arr), np.max(arr)`   | Min / Max                   |
|           | Linear Algebra     | `np.dot(A,B)`                | Matrix multiplication       |
|           |                    | `np.transpose(A)`            | Transpose                   |
|           |                    | `np.linalg.inv(A)`           | Inverse of matrix           |
|           |                    | `np.linalg.det(A)`           | Determinant                 |
|           | Random Numbers     | `np.random.rand(3)`          | Random numbers 0–1          |
|           |                    | `np.random.randint(0,10,5)`  | Random integers 0–9         |
|           |                    | `np.random.normal(0,1,5)`    | Normal distribution samples |
|           | Indexing / Slicing | `arr[0], arr[-1]`            | Access elements             |
|           |                    | `arr[1:4], arr[arr>5]`       | Slice or conditional filter |

-------------------------

Very customizable
Base for other libraries like Seaborn

----------------------------------

|                |                 | `plt.bar(x, y)`                      | Bar chart             |

# 4. Seaborn

Purpose: Make beautiful and complex plots easily using pandas data.

---------------------------

# Why we need it:
Matplotlib can do everything, but Seaborn makes complex plots simple, especially for statistics. Example: violin plots, heatmaps, pairplots. Works directly with pandas DataFrame.

-------------------------
# Key feature:
Stylish default themes
Easy for data exploration

-----------------------------------------------
| Library     | Category             | Function / Command                       | Description / Formula        |
| ----------- | -------------------- | ---------------------------------------- | ---------------------------- |
| **Seaborn** | Distribution         | `sns.histplot(data)`                     | Histogram                    |
|             |                      | `sns.kdeplot(data)`                      | Density / KDE plot           |
|             | Categorical          | `sns.barplot(x, y)`, `sns.boxplot(x, y)` | Compare categories           |
|             | Scatter / Regression | `sns.scatterplot(x, y)`                  | Scatter plot                 |
|             |                      | `sns.regplot(x, y)`                      | Scatter with regression line |
|             | Heatmap              | `sns.heatmap(df.corr())`                 | Correlation heatmap          |
|             | Pairwise             | `sns.pairplot(df)`                       | Pairwise relationships       |
|             | Styling              | `sns.set_style('whitegrid')`             | Set plot style / theme       |



# **NumPy Interview Questions**
What is NumPy and why is it used?

-->Fast numerical computation library; supports arrays and vectorized operations.
Difference between Python list and NumPy array.
Lists are slow and flexible, arrays are fast and memory-efficient.

-----------------------------

How to create a NumPy array?

-->import numpy as np
arr = np.array([1, 2, 3])

------------------------------
What is vectorization in NumPy?

-->Performing operations on entire arrays without loops (faster than for-loops).

----------------------------------

How to find mean, median, or standard deviation using NumPy?

-->np.mean(arr)

-->np.median(arr)

-->np.std(arr)

# **pandas Interview Questions**

What is pandas used for?

-->Handling structured data with rows and columns, like CSV or Excel files.

-----------------------
Difference between Series and DataFrame.

-->Series → 1D labeled data

-->DataFrame → 2D table with rows & columns

------------------------------
How to read a CSV in pandas?

-->import pandas as pd

df = pd.read_csv("data.csv")

-----------------------------------------
How to handle missing data?

-->df.fillna() → fill missing values
-->df.dropna() → remove rows with missing values

------------------------------------------

How to filter rows in pandas?

df[df['Age'] > 25]

# **Matplotlib Interview Questions**

Why use Matplotlib?

-->For creating visualizations like line plots, scatter plots, bar charts.

---------------------------------------
How to plot a simple line graph?

-->import matplotlib.pyplot as plt
-->plt.plot([1,2,3], [4,5,6])
-->plt.show()

-------------------------------------
Difference between plt.plot() and plt.scatter():

-->plot() → continuous line graph
-->scatter() → points plotted individually

----------------------------------
How to add labels and title to a plot?

-->plt.title("My Graph")

-->plt.xlabel("X-axis")

-->plt.ylabel("Y-axis")

# **Seaborn Interview Questions**

Why use Seaborn instead of Matplotlib?

-->Easier to make attractive statistical plots, works directly with pandas.

------------------------------------
Common Seaborn plots?

-->sns.barplot(), sns.boxplot(), sns.heatmap(), sns.pairplot()

-----------------------

How to create a bar plot with Seaborn?

-->import seaborn as sns
-->sns.barplot(x='Name', y='Age', data=df)

----------------------------------
Difference between Seaborn and Matplotlib?

-->Matplotlib → low-level plotting (full control)
--Seaborn → high-level, prettier, easier for statistical data
