```python
import matplotlib.pyplot as plt
import random

num_rand_data_points = 7

plt.figure(figsize=(10, 5), facecolor='w', edgecolor='k')

x_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]
y_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]

plt.plot(
    x_vals,
    y_vals,
    label="Marks",
    color="red",
    linestyle="--",
    linewidth=2,
    marker='*',
    markersize=20,
    markerfacecolor="orange",
    markeredgecolor="orange"
)

plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title(f"{num_rand_data_points} random data points!")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# setup the data
(x1_vals, y1_vals) = ([1,2,3,4], [1, 4, 9, 16])
(x2_vals, y2_vals) = ([0,1,1.5,7], [1,4,12,16])
(x3_vals, y3_vals) = ([0,1,3,5], [3,4.5,7,18])

# plot the data
plt.figure(figsize=(10, 5))

plt.plot(
    x1_vals, y1_vals,
    label="Red",
    color="red",
    marker='o', markersize=10
)

plt.plot(
    x2_vals, y2_vals,
    label="Green",
    color="green",
    marker='^', markersize=10
)

plt.plot(
    x3_vals, y3_vals,
    label="Blue",
    color="blue",
    marker='$❤$', markersize=10
)

plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C','D']
y = [5, 7, 3,2]

plt.figure(figsize=(6,3))

plt.bar(
    x, y,
    hatch="///",
    edgecolor="BLACK",
    linewidth=2.5,
    color="WHITE",
    bottom=4
)
plt.gca().set_facecolor("BLACK")
plt.minorticks_on()
plt.title("Bar Chart")
plt.xlabel("This is my X label name")
plt.ylabel("This is my Y label name")
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D', 'E']
y = [5, 7, 3, 8, 2]

plt.barh(
    x, y,
    color="blue",
    hatch="./",
    edgecolor="white",
    label="Blue"
)

plt.gca().set_facecolor("black")
plt.minorticks_on()

plt.title("Horizontal Bar Chart")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%' ,
    shadow=True,
    startangle=90,
    counterclock=False,
)

plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]
colors = ['red', 'blue', 'green', 'orange']

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'width': 0.5}
)

plt.title("Donut Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
days = [1, 2, 3, 4, 5]
apples = [3, 4, 2, 5, 4]
bananas = [1, 2, 3, 2, 1]
cherries = [2, 1, 2, 3, 2]

plt.stackplot(days, apples, bananas, cherries, labels=['Apples', 'Bananas', 'Cherries'], colors=['red', 'yellow', 'pink'])
plt.legend(loc='upper left')
plt.title('Fruit Sales Over Days')
plt.xlabel('Day')
plt.ylabel('Quantity')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C', 'D']
MATHS = [50, 60, 70, 80]
SCIENCE = [60, 70, 80, 90]
ENGLISH = [70, 80, 90, 100]

# Subplot 1: Line Plot (Maths)
plt.subplot(2, 2, 1)
plt.plot(students, MATHS, marker='o')
plt.title('Maths Marks')
plt.xlabel('Students')
plt.ylabel('Marks')

# Subplot 2: Bar Chart (Science)
plt.subplot(2, 2, 2)
plt.bar(students, SCIENCE)
plt.title('Science Marks')

# Subplot 3: Histogram (English)
plt.subplot(2, 2, 3)
plt.hist(ENGLISH, bins=5, color='lightblue', edgecolor='black')
plt.title("English Marks Distribution")
plt.xlabel('Marks')
plt.ylabel('Frequency')

# Subplot 4: Scatter Plot (Maths vs Science)
plt.subplot(2, 2, 4)
plt.scatter(MATHS, SCIENCE, color='green')
plt.title('Maths vs Science')
plt.xlabel('Maths')
plt.ylabel('Science')

plt.tight_layout()
plt.show()
```
