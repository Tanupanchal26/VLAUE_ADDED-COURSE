```python
x_values = [10,20,30,40]
plt.plot(
    x_values,
    label="Marks",
    color="Purple",
    linestyle=":",
    linewidth=2,
    marker='*',
    markersize=10
)
plt.ylabel("Y Label")
plt.xlabel("X Label")
plt.title("Graph")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

num_rand_data_points = 7

plt.figure(figsize=(10, 5), facecolor='w', edgecolor='k')

x_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]
y_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]

plt.plot(
    x_vals,
    y_vals,
    label="Marks",
    color="red",
    linestyle="--",
    linewidth=2,
    marker='*',
    markersize=20,
    markerfacecolor="orange",
    markeredgecolor="orange"
)

plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title(f"{num_rand_data_points} random data points!")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# setup the data
(x1_vals, y1_vals) = ([1,2,3,4], [1, 4, 9, 16])
(x2_vals, y2_vals) = ([0,1,1.5,7], [1,4,12,16])
(x3_vals, y3_vals) = ([0,1,3,5], [3,4.5,7,18])

# plot the data
plt.figure(figsize=(10, 5))

plt.plot(
    x1_vals, y1_vals,
    label="Red",
    color="red",
    marker='o', markersize=10
)

plt.plot(
    x2_vals, y2_vals,
    label="Green",
    color="green",
    marker='^', markersize=10
)

plt.plot(
    x3_vals, y3_vals,
    label="Blue",
    color="blue",
    marker='$❤$', markersize=10
)

plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
