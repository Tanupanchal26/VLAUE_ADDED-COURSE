```python
s = "Hello World"

print("String:", s)
print("First char:", s[0])
print("Last char:", s[-1])
print("Slice (0:5):", s[0:5])

# Iteration
for ch in s:
    print(ch, end=" ")
```

```python
a = "Hello"
b = "World"

print(a + " " + b)   # concatenation
print(a * 3)         # repetition
```

```python
s = "hello world"

print(s.upper())      # HELLO WORLD
print(s.lower())      # hello world
print(s.title())      # Hello World
print(s.swapcase())   # HELLO WORLD (reverse case)
```

```python
s = "hello world"

print(s.find("world"))      # 6
print(s.index("hello"))     # 0
print(s.startswith("he"))   # True
print(s.endswith("ld"))     # True
```

```python
name = "Tanya"
age = 18

# format()
print("My name is {} and I am {}".format(name, age))

# f-string
print(f"My name is {name} and I am {age}")

# % formatting
print("My name is %s and I am %d" % (name, age))
```

```python
s = "   TANYA   "

print(s.strip())
print(s.lstrip())
print(s.rstrip())

# padding
print(s.strip().center(20, '*'))
print(s.strip().ljust(20, '-'))
print(s.strip().rjust(20, '-'))
```

```python
s = "apple,banana,grape"

lst = s.split(",")
print(lst)

joined = "-".join(lst)
print(joined)
```

```python
s = "Tanya"

print("Count of a:", s.count("a"))
print("Count of n:", s.count("n"))
```

```python
s1 = "abc"
s2 = "123"
s3 = "abc123"
s4 = "   "

print(s1.isalpha())     # True
print(s2.isdigit())     # True
print(s3.isalnum())     # True
print(s4.isspace())     # True
```
